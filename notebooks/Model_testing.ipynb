```python
import pickle

with open("best_xgboost_model.pkl", "rb") as file:
    xgb_model = pickle.load(file)

with open("model_features.pkl", "rb") as file:
    model_features = pickle.load(file)

print("Model Loaded Successfully!")

```

```python
import pandas as pd 
import numpy as np
from xgboost import XGBRegressor
```

```python
sample_student = {
    'Age': 22,
    'Hours_Studied': 5,
    'Attendance': 92,
    'Sleep_Hours': 7,
    'Stress_Level': 3,
    'Screen_Time': 4.5,
    'Previous_GPA': 3.68,
    'Tutoring_Sessions_Per_Week': 2,
    'Exam_Anxiety_Score': 4,

    'Gender_Male': 1,
    'Gender_Non-Binary': 0,

    'Study_Method_Offline': 0,
    'Study_Method_Online': 0,

    'Diet_Quality_Good': 1,
    'Diet_Quality_Poor': 0,

    'Internet_Quality_Excellent': 0,
    'Internet_Quality_Good': 1,
    'Internet_Quality_Poor': 0,

    'Extracurricular_Yes': 1,
    'Part_Time_Job_Yes': 0,

    'Family_Income_Level_Low': 0,
    'Family_Income_Level_Middle': 0
}
```

```python
sample_df = pd.DataFrame([sample_student])
```

```python
sample_df["Study_Efficiency"] = (
    sample_df["Hours_Studied"] / (sample_df["Stress_Level"] + 1)
)

sample_df["Sleep_Study_Balance"] = (
    sample_df["Sleep_Hours"] * sample_df["Hours_Studied"]
)

sample_df["Anxiety_Attendance"] = (
    sample_df["Exam_Anxiety_Score"] * sample_df["Attendance"]
)

sample_df["Academic_Consistency"] = (
    sample_df["Previous_GPA"] / sample_df["Attendance"]
)

sample_df["Productivity_Score"] = (
    sample_df["Hours_Studied"] + sample_df["Attendance"]/10 + sample_df["Sleep_Hours"]
)

sample_df["Stress_per_Study_Hour"] = (
    sample_df["Stress_Level"] / (sample_df["Hours_Studied"] + 1)
)
```

```python
sample_df = sample_df[model_features]
```

```python
pred = xgb_model.predict(sample_df)

print("Predicted Final Score:", pred[0])
```
